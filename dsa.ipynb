```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(root.val, end=" ")
        inorder_traversal(root.right)

def preorder_traversal(root):
    if root:
        print(root.val, end=" ")
        preorder_traversal(root.left)
        preorder_traversal(root.right)

def postorder_traversal(root):
    if root:
        postorder_traversal(root.left)
        postorder_traversal(root.right)
        print(root.val, end=" ")

# Create a binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

print("Inorder Traversal:")
inorder_traversal(root)
print("\n")

print("Preorder Traversal:")
preorder_traversal(root)
print("\n")

print("Postorder Traversal:")
postorder_traversal(root)
print("\n")
```

```python

import matplotlib.pyplot as plt
import numpy as np

class TreeNode:
    def __init__(self, key, name):
        self.left = None
        self.right = None
        self.val = key
        self.name = name

def plot_tree(root, x=0, y=0, branch_width=2, node_size=300, level=1):
    if root:
        if root.left:
            x_left = x - 2 ** (5 - level)
            y_left = y - 1
            plt.plot([x, x_left], [y, y_left], 'k-', lw=branch_width)
            plot_tree(root.left, x_left, y_left, branch_width, node_size, level + 1)
        if root.right:
            x_right = x + 2 ** (5 - level)
            y_right = y - 1
            plt.plot([x, x_right], [y, y_right], 'k-', lw=branch_width)
            plot_tree(root.right, x_right, y_right, branch_width, node_size, level + 1)
        plt.scatter(x, y, s=node_size, color='r', zorder=100)
        plt.text(x, y, str(root.val), fontsize=12, ha='center', va='center', color='black')
        plt.text(x, y + 0.2, root.name, fontsize=10, ha='center', va='center', color='black')

# Create a binary tree with named nodes
root = TreeNode(1, "A")
root.left = TreeNode(2, "B")
root.right = TreeNode(3, "C")
root.left.left = TreeNode(4, "D")
root.left.right = TreeNode(5, "E")

# Plot the binary tree
plt.figure(figsize=(10, 6))
plot_tree(root)
plt.axis('off')
plt.show()
```
